```python
import pandas as pd
df=pd.read_csv('AMZN.csv')
df
```

```python
df=df.drop('Open',axis=1)
df=df.drop('Adj Close',axis=1)
df
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df.dtypes
```

```python
df=df.set_index('Date')
df
```

```python
df['Middle']=(df['High']+df['Low'])/2
df['HighNextDay']=df['High'].shift(-1)
df=df.dropna()
df
```

```python
def f(x):
    r='Y'
    if x['HighNextDay']<=x['High']:        
        r='N'
    return r
df['Result']=df.apply(f,axis=1)
df
```

```python
df.to_csv('AMZN_preprocess.csv')
```
