```python
import pandas as pd
import ta #conda install -c conda-forge ta
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('AMZN_preprocess.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df.dtypes
df=df.set_index('Date')
df
```

```python
#Create the Simple Moving Average (SMA)
def SMA(data,period=30,column='Middle'):
    return data[column].rolling(window=period).mean()

#Create the Exponential Moving Average (EMA)
def EMA(data,period=20,column='Middle'):
    return data[column].ewm(span=period,adjust=False).mean()
```

```python
#ADX,-DI,+DI
period=7
df['ADX']=ta.trend.ADXIndicator(df.High,df.Low,df.Close,window=period).adx()
df['+DI']=ta.trend.ADXIndicator(df.High,df.Low,df.Close,window=period).adx_pos()
df['-DI']=ta.trend.ADXIndicator(df.High,df.Low,df.Close,window=period).adx_neg()
df
```

-The ADX identifies a strong trend when the ADX is over 25 and a weak trend when the ADX is below 20.
- ADX>=25 Strong: 1
- 20<ADX<25: Neutral: 0
- ADX<20: Weak: -1

```python
df['ADX_Level']=np.where(df['ADX']>=25, 1, np.where(df['ADX']<20, -1, 0))
df
```

```python
-The ADX identifies a strong trend when the ADX is over 25 and a weak trend when the ADX is below 20.
- '+DI'>'-DI': Up trend  :  1
- '+DI'='-DI': Same      :  0
- '+DI'<'-DI': Down trend: -1
```

```python
df['ADX_Trend']=np.where(df['+DI']>df['-DI'], 1, np.where(df['+DI']<df['-DI'], -1, 0))
df
```

-On-Balance Volume (OBV)
-Volume itself is a valuable indicator, and on-balance volume (OBV) takes a significant amount of volume information and compiles it into a single one-line indicator. The indicator measures cumulative buying and selling pressure by adding the volume on "up" days and subtracting volume on "down" days.
-Ideally, the volume should confirm trends. A rising price should be accompanied by a rising OBV; a falling price should be accompanied by a falling OBV. 
-However, OBV need to identify trend by eyes, so we don't use it;

-Instead, we use Chaikin Forex indicator
-when the CMF volume forex indicator crosses above zero, itâ€™s seen as a buy signal. Conversely, when the CMF volume indicator crosses below zero, itâ€™s seen as a sell signal.

```python
df['CMF']=ta.volume.chaikin_money_flow(df.High,df.Low,df.Close,df.Volume,window=21)
df
```

```python
df=df.drop(columns=['HighNextDay','ADX','+DI','-DI',],axis=1)
df
```
